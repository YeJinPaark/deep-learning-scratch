### 8. 딥러닝

- `딥러닝`: 층을 깊게 한 심층 신경망

##### 8.1 더 깊게

- 앞의 내용들을 활용하여 심층 신경망 구현
    $\to$ MNIST 데이터셋의 손글씨 숫자 인식

##### 8.1.1 더 깊은 신경망

- `합성곱 계층`: 3*3 크기의 작은 필터  
    $\to$ 층이 깊어지면서 채널 수가 더 늘어남  
    (16 $\to$ 16 $\to$ 32 $\to$ 64 $\to$ 64)

- `풀링 계층`: 중간 데이터의 공간 크기를 점차 축소시킴  
- `완전연결 계층`(마지막 단): 드롭아웃 계층 사용

- `가중치 초깃값`: He 초깃값
- `가중치 매개변수 갱신`: Adam
- `활성화 함수`: ReLU

    $\to$ 신경망의 정확도: 99.38%  
$\equiv$ 신경망이 잘못 인식하는 이미지의 경우: 인간도 판단하기 어려운 이미지

##### 8.1.2 정확도를 높이려면

- `What is the class of this image?`: 다양한 데이터셋을 대상으로 논문에서 발표한 기법들의 정확도 순위  
$\to$ CNN을 기초로 한 기법들이 상위권 차지

- 정확도 개선에 영향: 앙상블 학습, 학습률 감소, 데이터 확장 등  
- `데이터 확장(data augmentation)`: 입력 이미지(훈련 이미지)를 알고리즘을 통해 '인위적'으로 확장  
    - ex) 입력 이미지를 `회전`하거나 `세로로 이동`하는 등의 미세한 변화 부여  
    - `crop`: 이미지 일부를 잘라냄  
    - `flip`: 좌우 반전, 이미지의 대칭성을 고려하지 않아도 되는 경우만 가능   
    - 발기 등의 외형 변화  
    - 확대, 축소 등의 스케일 변화


##### 8.1.3 깊게 하는 이유

- `층을 깊게 하는 것`  

    (1) `신경망의 매개변수 수가 줄어듬`  
    $\equiv$ 층을 깊게 한 신경망이 깊지 않은 경우보다 적은 매개변수로 같은 혹은 이상의 표현력 달성 가능

    - 출력 데이터의 각 노드가 입력 데이터의 어느 영역으로부터 계산되었는가  

    - ex) 5*5 합성곱 연산 1회(매개변수 5*5=25개)  
        $\equiv$ 3*3 합성곱 연산 2회 수행해 대체(매개변수 2*3*3=18개)

        $\to$ 층을 반복할수록 매개변수 수가 적어짐  
        $\equiv$ 개수의 차이가 층이 깊어질수록 커짐

        - ex) 3*3 합성곱 연산 3회 반복 시, 매개변수가 모두 27개  
            cf) 같은 크기의 영역을 1회의 합성곱 연산으로 하기 위해서는 7*7=49개의 매개변수 필요  

    - `수용 영역(receptive field)`: 뉴런에 변화를 일으키는 국소적인 공간 영역  
    $\to$ 작은 필터를 겹쳐 신경망을 깊게 하여, 매개변수 수를 줄여 넓은 수용영역을 소화

    (2) `학습의 효율성`  
    - 정보를 계층적으로 추출하고 있음
    $\equiv$ 층을 깊게 함으로써 학습 데이터의 양을 줄여, 학습을 고속으로 수행할 수 있음  
    $\equiv$ 즉, 한 번에 학습해야 할 데이터가 적어짐(학습해야 할 문제를 단순한 문제로 대체할 수 있음)

    (3) `정보를 계층적으로 전달`할 수 있음  
    - ex) 에지를 추출한 층의 다음 층은 에지 정보를 활용할 수 있음
    $\equiv$ 층을 깊이 할 수록, 더욱 더 고도의 패턴을 효과적으로 학습할 것이라 기대
