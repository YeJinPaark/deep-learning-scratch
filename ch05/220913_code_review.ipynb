5. 오차역전파법
    cf) 수치미분: 단순하고 구현하기 쉬움, but 오랜 연산 시간

5.1 계산 그래프(computational graph)
    - 그래프: 복수의 node와 egde로 표현

5.1.1 계산 그래프
    - node를 원으로 표기, 내부에 연산 내용 명시
    - 순전파(forward propagation): 왼쪽에서 오른쪽으로 계산 진행
    - 역전파(backward propagation): 미분 연산에 중요

5.1.2 국소적 계산
    - 국소적: 자신과 직접 관계된 작은 범위
    - 국소적 계산: 전체에서 어떤 일이 벌어지든 상관없이 자신과 관계된 정보만으로 다음 결과를 출력할 수 있음
    - 계산 그래프는 국소적 계산에 집중 == 복잡한 자동차 조립 중 조립 라인 작업으로 분업

5.1.3 왜 계산 그래프인가
    - 이점
        1. 국소적 계산으로 각 노드에서 단순한 계산에 집중해 문제를 단순화 할 수 있음
        2. 중간 계산 결과를 모두 보관할 수 있음
        3. 역전파를 통해 미분을 효율적으로 계산할 수 있음

5.2 연쇄법칙(chain rule)

5.2.1 계산 그래프의 역전파
    - 극소적인 미분을 상류에서 전달된 값에 곱해 앞쪽 노드로 전달

5.2.2 연쇄법칙
    - 합성 함수: 여러 함수로 구성된 함수
    - 연쇄 법칙의 원리: 합성 함수의 미분은 합성 함수를 구성하는 각 함수의 미분의 곱으로 나타낼 수 있음

5.2.3 연쇄법칙과 계산 그래프
    - 역전파: 오른쪽에서 왼쪽으로 신호 전파

5.3 역전파

5.3.1 덧셈 노드의 역전파
    - 입력된 값을 그대로 다음 노드로 전달
    - 순방향 입력 신호의 값이 필요하지 않음
    ex) z=x+y, 덧셈 노드 기준
        - aZ/ax=1-> aL/ax*1
        - aZ/ay=1-> aL/ay*1

5.3.2 곱셈 노드의 역전파
    - 상류의 값에 순전파 때의 입력 신호들을 '서로 바꾼 값'을 곱해서 하류로 전달
    - 순방향 입력 신호의 값이 필요
    ex) z=xy, 곱셈 노드 기준
        - aZ/ax=y-> aL/ax*y
        - aZ/ay=x-> aL/ay*x
