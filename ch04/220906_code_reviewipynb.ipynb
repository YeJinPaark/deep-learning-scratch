4.1 
    - 학습: 훈련 데이터로부터 가중치 매개변수의 최적값을 자동으로 획득하는 것
    - 학습 목표: 손실 함수의 결괏값을 가장 작게 만드는 가중치 매개변수를 찾는 것
    - 경사법: 함수의 기울기 활용해 손실 함수 값 최소화

4.1.1 데이터 주도 학습
    - 특징을 추출하고, 특징의 패턴을 기계학습 기술로 학습
    - 특징: 입력 데이터에서 본질적인 데이터를 정확하게 추출할 수있도록 설게된 변환기
    ex) 이미지의 특징은 벡터로 기술, 컴퓨터 비전에서는 SIFT, sURF, HOG 등의 특징 사용-> SVM, KNN 등의 지도 학습 방식의 분류 기법으로 학습

    - 기계 학습
        1. 사람이 생각한 특징(SIFT, HOG 등) + 기계 학습(SVM, KNN 등)
        2. 신경망(딥러닝) == 종단간 기계 학습(입력에서 목표한 출력을 얻는다)

4.1.2 훈련 데이터와 시험 데이터
    - 분리 이유: 범용 능력을 제대로 평가하기 위함
        * 범용 능력: 아직 보지 못한 데이터로도(훈련 데이터 제외) 문제를 올바르게 풀어내는 능력
    - overfitting: 한 데이터셋에만 지나치게 최적화된 상태
